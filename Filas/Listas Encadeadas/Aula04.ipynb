# IMPLEMENTAÇÃO DE UMA LISTA ENCADEADA

# Nó da Lista

```python
class Node:
    def __init__(self, qtd, preco, descricao):
        self.qtd = qtd
        self.preco = preco
        self.descricao = descricao
        self.next = None
```

# Lista

```python
class linked_list:
    def __init__(self):
        self.head = None

# Inserção de Elementos na Lista Encadeada
    def inserir_no_inicio(self, qtd, preco, descricao):
        new_node = Node(qtd, preco, descricao)
        new_node.next = self.head
        self.head = new_node

    def inserir_no_fim(self, qtd, preco, descricao):
        new_node = Node(qtd, preco, descricao)
        if self.head is None:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node

    def inserir_em_posicao(self, qtd, preco, descricao, posicao):
        new_node = Node(qtd, preco, descricao)
        if posicao == 0:
            self.inserir_no_inicio(qtd, preco, descricao)
            return
        else:
            current_node = self.head
            for i in range(posicao - 1):
                if current_node is None:
                    print("Posição inválida")
                    return
                current_node = current_node.next
            
            if current_node is None:
                print("Posição inválida")
                return
            
            new_node.next = current_node.next
            current_node.next = new_node

# Remoção de Elementos da Lista Encadeada
    def remover_do_inicio(self):
        if self.head is None:
            print("A Lista está vazia")
            return
        self.head = self.head.next

    def remover_de_posicao(self, posicao):
        if self.head is None:
            print("A Lista está vazia")
            return
        if posicao == 0:
            self.remover_do_inicio()
            return
        current_node = self.head
        for i in range (posicao - 1):
            if current_node is None:
                print("Posição inválida")
                return
            current_node = current_node.next
        if current_node is None or current_node.next is None:
            print("Posição inválida")
            return
        current_node.next = current_node.next.next

    def remover_do_fim(self):
        if self.head is None:
            print("A Lista está vazia")
            return
        if self.head.next is None:
            self.head = None
            return
        last = self.head
        while last.next.next:
            last = last.next
        last.next = None

# Acesso a elementos em posição específica
    def acessar_posicao(self, posicao):
        if self.head is None:
            print("A Lista está vazia")
            return
        current_node = self.head
        for i in range(posicao):
            if current_node is None:
                print("Posição inválida")
                return
            current_node = current_node.next
        if current_node is None:
            print("Posição inválida")
            return
        else:
            # return f"Quantidade: {current_node.qtd}, Preco: {current_node.preco}, Descricao: {current_node.descricao}"
            return current_node.qtd, current_node.preco, current_node.descricao
        
# Buscar elementos
    def buscar_posicao_elemento(self, qtd, preco, descricao):
        current_node = self.head
        posicao = 0 
        while current_node:
            # if current_node.qtd == qtd and current_node.preco == preco and current_node.descricao == descricao:
            if (current_node.qtd, current_node.preco, current_node.descricao) == (qtd, preco, descricao):
                return posicao
            current_node = current_node.next
            posicao += 1
        return -1
    

# Atualizar elementos
    def atualizar_elemento_posicao(self, posicao, new_qtd, new_preco, new_descricao):
        if self.head is None:
            print("A Lista está vazia")
            return
        current_node = self.head
        for i in range(posicao):
            if current_node is None:
                print("Posição inválida")
                return
            current_node = current_node.next
        if current_node is None:
            print("Posição inválida")
            return
        current_node.qtd = new_qtd
        current_node.preco = new_preco
        current_node.descricao = new_descricao

# Remover elementos duplicados
    def remover_duplicados(self):
        if self.head is None:
            print("A Lista está vazia")
            return
        current_node = self.head
        next_node = current_node.next
        while current_node:
            next_node = current_node
            while next_node and next_node.next:
                if (current_node.qtd, current_node.preco, current_node.descricao) == (next_node.next.qtd, next_node.next.preco, next_node.next.descricao):
                    next_node.next = next_node.next.next
                else:
                    next_node = next_node.next
            current_node = current_node.next

# Exibição da Lista Encadeada
    def exibir_lista(self):
        current_node = self.head
        while current_node:
            print(f"Quantidade: {current_node.qtd}, Preco: {current_node.preco}, Descricao: {current_node.descricao}")
            current_node = current_node.next
            
```
